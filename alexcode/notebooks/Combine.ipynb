```python
for i in range(subm1.shape[0]):
    mu_1 = sum((np.array(range(599)) + 0.5) * np.diff(subm1.iloc[i, 1:601]))
    mu_2 = sum((np.array(range(599)) + 0.5) * np.diff(subm2.iloc[i, 1:601]))
    s_1 = sum((np.array(list(map(lambda x: x * x, np.array(range(599)) + 0.5)))) * np.diff(subm1.iloc[i, 1:601])) - mu_1 * mu_1
    s_2 = sum((np.array(list(map(lambda x: x * x, np.array(range(599)) + 0.5)))) * np.diff(subm2.iloc[i, 1:601])) - mu_2 * mu_2
    a = norm.cdf(np.linspace(0, 599, 600), (mu_1 + mu_2) / 2, np.sqrt((s_1 + s_2)/4))
    subm1.iloc[i, 1:601] = a
    

subm1.to_csv('../submissions/sub_combo_matstat.csv', index = False)
```
